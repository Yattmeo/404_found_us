```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv('truncated_dataset.csv')
data.head()
```

```python
df_1 = data[['date','transaction_id','card_brand','card_brand','card_type','merchant_id', 'mcc','mcc_description', 'amount']]
df_1.info()
```

```python
most_transactions = df_1['merchant_id'].value_counts().idxmax()
print(f'The merchant with the highest number of transactions is: {most_transactions}')

merchant_mt_df = df_1[df_1['merchant_id'] == most_transactions]
```

```python
merchant_mt_df.sort_values(by='date', inplace=True)
merchant_mt_df.head()
```

```python
m2 = merchant_mt_df[['date', 'transaction_id', 'amount']]

m2.head()
```

```python
m2['month'] = pd.to_datetime(m2['date']).dt.to_period('M')

print(m2.head())
print('---')
print(m2.tail())
```
